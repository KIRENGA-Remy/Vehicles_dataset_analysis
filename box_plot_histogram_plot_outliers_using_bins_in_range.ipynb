```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

file_path = 'vehicles.csv'


try:
    df = pd.read_csv(file_path)
    plt.figure(figsize=(20,3))
    sns.boxplot(data=df[['selling_price','proposed_purchase_price']],palette="Accent")
    plt.title("Box plot of selling price and purchase price",fontsize=15)
    
    plt.show()
    selling_bins = [2000, 11000, 24000, 33000]
    proposed_purchase_price_bins = [2000, 12000, 30000, 49000,56000]
    plt.figure(figsize=(23, 3))
    counts, bins, _ = plt.hist(df['selling_price'], bins=selling_bins, edgecolor='yellow')
    for count, bin_edge in zip(counts, bins[:-1]):
        plt.text(bin_edge + (bins[1] - bins[0]) / 2, count , f'{int(count)}', fontsize=10)
    plt.title("Selling price histogram")
    plt.show()
    plt.figure(figsize=(23, 3))
    counts, bins, _ = plt.hist(df["proposed_purchase_price"], bins=proposed_purchase_price_bins, edgecolor='green')
    for count, bin_edge in zip(counts, bins[:-1]):
        plt.text(bin_edge + (bins[1] - bins[0]) / 2, count, f'{int(count)}', fontsize=10)
    plt.title("Proposed Price histogram")
    plt.show()
except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")
except pd.errors.EmptyDataError:
    print("Error: The file is empty.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
